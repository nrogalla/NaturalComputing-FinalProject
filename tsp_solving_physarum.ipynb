## Slime Mould

The idea is the following:
- the first phase has the physarum grow and establish a network (like Tokyo).
- in the second phase, first, all connections that violate the TSP are removed (so that all nodes have max 2 connections), then a reduced graph is established by removing all fully connected cities, then the remaining connections are established by either randomly or via nearest neighbor adding connections to achieve full tsp solution
- in the third step the found tsp solution is used as the initialisation to a known tsp solver (right now sim annealing)
